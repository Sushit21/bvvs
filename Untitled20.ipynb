```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('books.csv')

# Count the number of books per genre
genre_counts = df['Genre'].value_counts()

# Plotting the bar graph
plt.figure(figsize=(12, 6))
genre_counts.plot(kind='bar', color='steelblue', edgecolor='black')

plt.title('Number of Books by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Books')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('books.csv')

# Drop rows with missing values in Genre or Height
df = df.dropna(subset=['Genre', 'Height'])

# Assign numeric values to genres for plotting
genre_codes = pd.factorize(df['Genre'])[0]

# Plotting
plt.figure(figsize=(12, 6))
plt.scatter(genre_codes, df['Height'], alpha=0.6, color='darkorange', edgecolors='black')

# Customizing x-axis with genre labels
plt.xticks(ticks=range(len(df['Genre'].unique())), labels=df['Genre'].unique(), rotation=45, ha='right')

plt.title('Book Height by Genre')
plt.xlabel('Genre')
plt.ylabel('Height (mm)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('books.csv')

# Drop missing values in the Height column
df = df.dropna(subset=['Height'])

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Height'], bins=20, color='mediumslateblue', edgecolor='black', alpha=0.75)

plt.title('Distribution of Book Heights')
plt.xlabel('Height (mm)')
plt.ylabel('Number of Books')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('books.csv')

# Drop rows with missing values in 'Genre' or 'Height'
df = df.dropna(subset=['Genre', 'Height'])

# Plotting the boxplot
plt.figure(figsize=(12, 6))
df.boxplot(column='Height', by='Genre', grid=False, patch_artist=True,
           boxprops=dict(facecolor='lightblue', color='blue'),
           medianprops=dict(color='red'),
           whiskerprops=dict(color='blue'),
           capprops=dict(color='blue'),
           flierprops=dict(marker='o', color='orange', alpha=0.5))

plt.title('Book Height Distribution by Genre')
plt.suptitle('')  # Removes default subtitle
plt.xlabel('Genre')
plt.ylabel('Height (mm)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('books.csv')

# Drop missing values in the column you're analyzing
df = df.dropna(subset=['Height'])

# Set the visual style
sns.set(style='whitegrid')

# Plot the density chart
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Height', fill=True, color='skyblue', linewidth=2)

plt.title('Density Plot of Book Heights')
plt.xlabel('Height (mm)')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```
